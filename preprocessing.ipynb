### Video name to vid length

```python
import os

vid2len={}
directory = os.fsencode('./e6691-bucket-images/')
for file in os.listdir(directory):
    filename = os.fsdecode(file)
    lenght=len(os.listdir(os.path.join(directory,file)))
    vid2len[filename]=lenght
vid2len
```

```python
def frame2label(name,frame,df):

    r = df.loc[(df['videoName'] == name) &(df['End_sec'] >frame) & (df['Start_sec'] <= frame)]
    return r[['class']].values[0][0]

#example
frame2label('RALIHR_surgeon01_fps01_0001',1,df)
```

```python
def vid2labelseq(name,df):
    print('processing', name)
    df_loc=df.loc[(df['videoName'] == name) ]
    length=vid2len[name]
    labels=[]
    for i in range(length):
        labels.append(frame2label(name,i,df_loc))
    return labels
#example
vid2labelseq('RALIHR_surgeon01_fps01_0001',df)
```

```python
vids2label={}
for file in df['videoName'].unique():
    vids2label[file]=vid2labelseq(file,df)
```

```python
import pickle

with open('./pickle/vids2label.p', 'wb') as fp:
    pickle.dump(vids2label, fp, protocol=pickle.HIGHEST_PROTOCOL)

with open('./pickle/vids2label.p', 'rb') as fp:
    vids2label = pickle.load(fp)
```

```python
frames2label={}

for vid in vids2label:
    for frame,label in enumerate(vids2label[vid]):
        frames2label[vid+'/frame'+str(frame)+'.jpg']=label
    print(vid)
frames2label
```
